```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
# スクレイピングするウェブサイトのURLを定義
url1 = 'https://in-note.com/h/words'
```

```python
# ウェブページのHTMLコンテンツを取得
response = requests.get(url1)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
elems = soup.find_all(href=re.compile("words"))
print(elems)
```

```python

pickup_links=[]
for elem in elems:
  if elem.contents != [] and elem.contents != ['単語一覧']:
    #print(elem.contents)
    #print(elem.attrs['href'])
    pickup_links.append('https://in-note.com'+elem.attrs['href'])

print(pickup_links)

```

```python
links=[]
for pickup_link in pickup_links:
  #Pickupページへ遷移しページの情報を取得
    pickup_res = requests.get(pickup_link)
    pickup_soup = BeautifulSoup(pickup_res.text, "html.parser")

    elems = pickup_soup.find_all(href=re.compile("words?"))
    for elem in elems:
      if elem.contents != [] and elem.contents != ['単語一覧']:
        links.append('https://in-note.com'+elem.attrs['href'])

print(links)
```

```python
text = []

for link in links:
    pickup_res = requests.get(link)
    pickup_soup = BeautifulSoup(pickup_res.text, "html.parser")

    word_lists = pickup_soup.find_all(class_=re.compile("word-list"))
    for word_list in word_lists:
        items = re.split(r'(?<=）)', word_list.get_text(strip=True))
        text.extend([item for item in items if item.strip()])
```

```python
print(text)
```

**・下は練習用**

```python
# 正規表現を使用してふりがなを抽出し、元のテキストと分ける
original = []
furigana = []
```

```python
for entry in text:
    original_text = re.sub(r'（.*?）', '', entry)
    furigana_match = re.search(r'（(.*?)）', entry)
    if furigana_match:
        furigana_text = furigana_match.group(1)
    else:
        furigana_text = ""
    original.append(original_text)
    furigana.append(furigana_text)
```

```python
print(original)
```

```python
# データフレームを作成
df = pd.DataFrame({
    'Original': original,
    'Furigana': furigana
})
```

```python
# CSVファイルとして保存
df.to_csv('output.csv', index=False)
```
